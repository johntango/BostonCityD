```python
import json
import urllib.request
import urllib.parse
import urllib.error
import pandas as pd 
```

```python
url = "https://pollysnips.s3.amazonaws.com/bostonEmployeeSalaries.json"
uh = urllib.request.urlopen(url)
data = uh.read().decode()
print('Retrieved', len(data), 'characters')
print(type(data))
print(data[:2])
# print the structure of the data

# load the data into a dictionary
info = json.loads(data)
myData = info["data"]

print(myData[0:5])
print("Total salary:"+  myData[0][18])
print("Total salary:"+  myData[1][18])
print(type(myData[0][18]))
```

```python
# store the data in a pandas dataframe
df = pd.DataFrame(myData)
five = df[18][:5]
print(type(five))


```

```python
# Note that Pandas DataFrames are indexed by column name(or number) first and then row number
# plot salaries as a pandas histogram
# convert the salary column to a float
salary = df[18].astype(float) # convert to float
salary.plot.hist(bins=20)
# print total of all salaries
print("Total salary:"+  str(salary.sum()))

```

```python
# plot out salaries above 200,000 as a pandas histogram
salary[salary > 200000].plot.hist(bins=20)
# print total of all salaries above 200,000
print("Total salary above 200,000:"+  str(salary[salary > 200000].sum()))
```
